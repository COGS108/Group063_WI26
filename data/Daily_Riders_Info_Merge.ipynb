```python
import pandas as pd
# "~/Desktop/MTA_Subway_Hourly_Ridership__2020-2024_20260222.csv"
daily_ridership = pd.read_csv("~/Desktop/MTA_Subway_Daily_Manhattan.csv")
st_cmplx_list = pd.read_csv("~/Desktop/MTA_Subway_Daily_Manhattan_Stations.csv")
```

```python
daily_ridership_plus_info = daily_ridership.merge(
    st_cmplx_list, 
    on='station_complex_id', 
    how='left'
)
```

```python
daily_ridership_plus_info
```

```python
# 1. Use regex to extract the two numbers inside the parentheses
# ([-\d.]+) matches digits, decimals, and negative signs
new_cols = daily_ridership_plus_info['Georeference'].str.extract(r'POINT \(([-\d.]+) ([-\d.]+)\)')

# 2. Assign to new columns and convert to floats
daily_ridership_plus_info['longitude'] = pd.to_numeric(new_cols[0])
daily_ridership_plus_info['latitude'] = pd.to_numeric(new_cols[1])

# 3. Optional: Drop the original column if no longer needed
daily_ridership_plus_info = daily_ridership_plus_info.drop(columns=['Georeference'])

daily_ridership_plus_info
```

```python
unique_ids = daily_ridership_plus_info.station_complex_id.unique()
expected_rows = len(unique_ids) * 365
print('The number of expected rows, based on the number of station_complex_ids: ', expected_rows)
print('The number of missing rows is: ', expected_rows - len(daily_ridership_plus_info))

# 1. Get the counts
# 2. Convert to DataFrame (reset_index turns the ID back into a column)
# 3. Rename columns to your preference
num_days_each_station = (
    daily_ridership_plus_info['station_complex_id']
    .value_counts()
    .reset_index()
)

num_days_each_station.columns = ['station_complex_id', 'op_days']
num_days_each_station['op_days'] = pd.to_numeric(num_days_each_station['op_days'])
```

```python
# 1. Create the Master Calendar as Timestamps (NOT .date)
all_dates = pd.date_range(start='2023-01-01', end='2023-12-31')

# 2. Get your incomplete IDs
incomplete_ids = num_days_each_station[num_days_each_station['op_days'] < 365]['station_complex_id']

# 3. Create the 'Perfect Grid'
master_grid = pd.MultiIndex.from_product(
    [incomplete_ids, all_dates], 
    names=['station_complex_id', 'date']
).to_frame(index=False)

# 4. FORCE TYPE ALIGNMENT
# Convert both 'date' columns to datetime64[ns] to be 100% sure
master_grid['date'] = pd.to_datetime(master_grid['date'])
daily_ridership_plus_info['date'] = pd.to_datetime(daily_ridership_plus_info['date'])

# Ensure IDs are same type (e.g., int)
master_grid['station_complex_id'] = master_grid['station_complex_id'].astype(int)
daily_ridership_plus_info['station_complex_id'] = daily_ridership_plus_info['station_complex_id'].astype(int)

# 5. Re-run the Merge
missing_days = master_grid.merge(
    daily_ridership_plus_info[['station_complex_id', 'date', 'ridership']], 
    on=['station_complex_id', 'date'], 
    how='left', 
    indicator=True
)

# 6. Isolate the missing ones
missing_days_final = missing_days[missing_days['_merge'] == 'left_only'].drop(columns=['_merge', 'ridership'])

print(f"Found {len(missing_days_final)} missing date/station pairs.")
```
