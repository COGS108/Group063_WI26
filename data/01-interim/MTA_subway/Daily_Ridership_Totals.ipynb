```python
import pandas as pd
# Oringinal Data (RAW) "MTA_Subway_Hourly_Ridership__2020-2024_20260222.csv"
daily_ridership = pd.read_csv("MTA_Subway_Daily_Manhattan.csv")
st_cmplx_list = pd.read_csv("MTA_Subway_Daily_Manhattan_Stations.csv")
```

```python
daily_ridership
```

```python
daily_ridership_totals = daily_ridership.groupby('date')['ridership'].sum().reset_index()
```

```python
daily_ridership_totals
```

```python
daily_ridership_totals.to_csv("MTA_Subway_Daily_Manhattan_Totals.csv", index=False)
```
