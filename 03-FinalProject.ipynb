## Background and Prior Work

## Background
Public transit agencies have pursued service expansions and frequency improvements to attract riders, yet many U.S. cities experienced stagnation or decline in ridership in the years leading up to COVID-19. A Congressional Research Service review notes that, despite major capital investments (including rail expansions), overall ridership fell in many large markets, and new rail lines do not necessarily increase total system ridership– sometimes shifting resources away from bus networks that carry a large share of trips.<sup><a href="#fn1" id="ref1">1</a></sup> At the same time, multiple competitive and behavioral forces changed the travel market: cheaper gasoline, rising fares, telework, and the growth of alternative modes like bikeshare and ride-hailing.<sup><a href="#fn1" id="ref1">1</a></sup>

## Prior work
Recent empirical work suggests that transit supply improvements can raise ridership, but the net outcome depends on the broader context. Erhardt et al. (2022) quantify ridership change across U.S. metropolitan areas pre-pandemic and find that while expanded transit service and land-use changes would have increased ridership, these gains were outweighed by other factors—especially ride-hailing along with higher fares, higher incomes, more teleworking, higher car ownership, and lower gas prices.<sup><a href="#fn3" id="ref3">3</a></sup> Complementing this, Graehler et al. (TRB 2019) document mode-specific ridership declines in major U.S. cities since the mid-2010s and enumerate commonly cited explanations including service cuts, deferred-maintenance reliability issues, micro mobility growth, and Transportation Network Company(TNC) expansion(Uber/Lyft); their panel analysis finds that standard factors (service levels, gas price, auto ownership) do not fully explain the recent declines and reports negative associations between years-since-TNC entry and bus/heavy-rail ridership.<sup><a href="#fn2" id="ref2">2</a></sup>

## Motivation for this project
The New York City subway system is a critical part of its transportation with millions of people riding it daily. The increase in use of uber and lyft, with over half a million uber and lyft rides requested in New York City alone, people raise the question of whether these services are pulling riders away from the subway system. It's a well known fact that public transport systems, such as MTA, generally lose money by being in operation. With ride hailing becoming an increasing way that people transport through the city, it may increase that deficit. To further investigate Erhardt et al. and Graehler et al. research claims, with our analysis we would be able to really pinpoint where the ridership declines and ridehailing increases.

## Hypothesis


After accounting for trends (seasonality, weather, and gas prices), stations with high surrounding ride hailing activity will have a lower number of riders per day. We anticipate that during more intense weather conditions (heat, cold, rain, and snow), subway stations will experience less ridership and ride hailing activity will increase accordingly. Furthermore, we anticipate as gas prices increase, ridesharing prices will follow suit, lowering their ridership, and increasing subway ridership.

The hypothesized thresholds are as follows:

- Temperature (on days when it is not raining or snowing):
    - Lower threshold: 50°F (10°C) (temperatures below this will lead to a decrease in subway ridership)
    - Upper threshold: 80°F (27°C) (temperatures above this will lead to a decrease in subway ridership)
- Rain: any amount of rain will lead to a decrease in ridership
- Snow: any amount of snow will lead to a decrease in ridership
- Gas Prices: $3.50 (gas prices exceeding this amount will lead to increased subway ridership)

### MTA Subway Hourly Ridership (2023)

The main variable that we are interested in is ridership, which is the total number of riders that entered a subway recorded through the New York City payment system card. This ridership value includes transfers, those who use a free bus-to-subway or free out-of-network transportation. For example, Time Square station complexes could get thousands of entries while outskirts complexes could get significantly less. The MTA Subway Hourly Ridership would allow us to see the times of peak ridership along with the lows of ridership.

A crucial variable that this dataset provides is transit_timestamps that are recorded to each hour, so a 1:37pm tie would be recorded as a 1:00pm transite_timestamp. The format is YYYY-MM-DDTHH:MM:SS.SSS, however the “:MM:SS.SSS” is always :00:00.000. As these are strings, this format allows us to use pd.to_datetime(), a more usable form.
The location provided in the dataset is critical when analyzing other factors in comparison. Station_complex_id are the unique identifiers for station complexes while station_complex are the more human recognizable names of stations and these stations can support many subway lines. Another location specifying variable is latitude,  longitude, and borough which also be critical for when compared to other factors.


One of the biggest concerns we have is the massive size of the datasets that we will be handling. To manage this we chose to use the "Query data" action on the website to reduce the dataset to the year 2023 and the specific variables we need. This resulted in a 3 GB CSV file which we processed locally to further reduce the size before adding the dataset to our Github repo.

Another concern is with ridership counts. Due to the ridership counts being counted based on fare payment transactions, those who evade subway fare will not be counted. So systemically, the ridership number will be underreported. MTA reports that subway fare evasion is around 10% which has decreased from the year before (https://cbcny.org/research/no-fare). Due to the volatile rate of fare evasion, it may bias the trend.

````python
ridehail_df = pd.read_csv("data/01-interim/TLC_ridehail/ridehailing_daily_ridership.csv")
#print(ridehail_df.columns)
# Convert dates
ridehail_df['date'] = pd.to_datetime(ridehail_df['date'])

# Add day of week columns
ridehail_df['day_of_week'] = ridehail_df['date'].dt.day_name()

# Add weekend column (using dayofweek numbers: 5=Saturday, 6=Sunday)
ridehail_df['weekend'] = ridehail_df['date'].dt.dayofweek.isin([5, 6])
ridehail_total = ridehail_df.groupby('date', as_index=False)['trip_count'].sum()

# Split data into weekday and weekend
weekday_data = ridehail_df[~ridehail_df['weekend']]
weekend_data = ridehail_df[ridehail_df['weekend']]

# Aggregate to daily totals
weekday_total = weekday_data.groupby('date', as_index=False)['trip_count'].sum()
weekend_total = weekend_data.groupby('date', as_index=False)['trip_count'].sum()

# Merge back day_of_week for printing purposes
weekday_total = weekday_total.merge(weekday_data[['date', 'day_of_week']].drop_duplicates(), on='date', how='left')
weekend_total = weekend_total.merge(weekend_data[['date', 'day_of_week']].drop_duplicates(), on='date', how='left')

# Calculate statistics for weekdays
weekday_mean = weekday_total['trip_count'].mean()
weekday_std = weekday_total['trip_count'].std()
weekday_upper_threshold = weekday_mean + (3 * weekday_std)
weekday_lower_threshold = weekday_mean - (3 * weekday_std)
weekday_upper_outliers = weekday_total[weekday_total['trip_count'] > weekday_upper_threshold]
weekday_lower_outliers = weekday_total[weekday_total['trip_count'] < weekday_lower_threshold]

# Calculate statistics for weekends
weekend_mean = weekend_total['trip_count'].mean()
weekend_std = weekend_total['trip_count'].std()
weekend_upper_threshold = weekend_mean + (3 * weekend_std)
weekend_lower_threshold = weekend_mean - (3 * weekend_std)
weekend_upper_outliers = weekend_total[weekend_total['trip_count'] > weekend_upper_threshold]
weekend_lower_outliers = weekend_total[weekend_total['trip_count'] < weekend_lower_threshold]

# Print weekday outliers
print(f"\nWEEKDAY - HIGH OUTLIERS (above +3σ): {len(weekday_upper_outliers)} days found")
print("-" * 50)
if len(weekday_upper_outliers) > 0:
    for idx, row in weekday_upper_outliers.iterrows():
        print(f"Date: {row['date'].strftime('%Y-%m-%d')}, "
            f"trip_count: {row['trip_count']:.0f}, "
            f"Day: {row['day_of_week']} "
            f"(+{(row['trip_count'] - weekday_mean)/weekday_std:.2f}σ)")
else:
    print("No weekday high outliers found above +3σ")

print(f"\nWEEKDAY - LOW OUTLIERS (below -3σ): {len(weekday_lower_outliers)} days found")
print("-" * 50)
if len(weekday_lower_outliers) > 0:
    for idx, row in weekday_lower_outliers.iterrows():
        print(f"Date: {row['date'].strftime('%Y-%m-%d')}, "
            f"trip_count: {row['trip_count']:.0f}, "
            f"Day: {row['day_of_week']} "
            f"({(row['trip_count'] - weekday_mean)/weekday_std:.2f}σ)")
else:
    print("No weekday low outliers found below -3σ")

# Print weekend statistics
print(f"\nWeekend Statistics:")
print(f"Mean trip_count (Weekends): {weekend_mean:.2f}")
print(f"Standard deviation: {weekend_std:.2f}")
print(f"Upper 3-sigma threshold (+3σ): {weekend_upper_threshold:.2f}")
print(f"Lower 3-sigma threshold (-3σ): {weekend_lower_threshold:.2f}")

print(f"\nWEEKEND - HIGH OUTLIERS (above +3σ): {len(weekend_upper_outliers)} days found")
print("-" * 50)
if len(weekend_upper_outliers) > 0:
    for idx, row in weekend_upper_outliers.iterrows():
        print(f"Date: {row['date'].strftime('%Y-%m-%d')}, "
            f"trip_count: {row['trip_count']:.0f}, "
            f"Day: {row['day_of_week']} "
            f"(+{(row['trip_count'] - weekend_mean)/weekend_std:.2f}σ)")
else:
    print("No weekend high outliers found above +3σ")

print(f"\nWEEKEND - LOW OUTLIERS (below -3σ): {len(weekend_lower_outliers)} days found")
print("-" * 50)
if len(weekend_lower_outliers) > 0:
    for idx, row in weekend_lower_outliers.iterrows():
        print(f"Date: {row['date'].strftime('%Y-%m-%d')}, "
            f"trip_count: {row['trip_count']:.0f}, "
            f"Day: {row['day_of_week']} "
            f"({(row['trip_count'] - weekend_mean)/weekend_std:.2f}σ)")
else:
    print("No weekend low outliers found below -3σ")
````

### Gas Price Dataset
The organization that published this dataset is the U.S. Energy Information Administration (EIA) which is an independent federal statistical agency that collects, analyzes, and disseminates independent and impartial energy information. This specific dataset contains weekly average retail prices for regular gasoline in the New York City area in dollars per gallon. The dataset has two variables: the week-ending date and the average retail price in USD per gallon. It covers data from 2000 to present year which means there are approximately 1,340 weekly observations. The prices that consumers pay range from a little more than a dollar to nearly five dollars per gallon.

The data reveals major price spikes that coincide with significant events, such as Hurricane Katrina, which hit oil-producing regions and caused a spike in prices. In 2008, oil prices were at an all-time high that could be attributed to financial speculation and coinciding with the US being amidst a great recession. The peak of 2022 can be suggested to be due to the post pandemic activities that occur after things start opening up after a shutdown. Overall, lower prices under 3.00 dollars per gallon tend to correspond with periods of economic downs while prices above 4.00 dollars per gallon indicate increasing global demands.

Because this dataset reports an area-level average for New York City, it smooths over station-to-station variation, differences across neighborhoods, and proximity to highways, giving a single representative price signal for the region. A drawback is that the EIA's NYC pricing area may not perfectly align with the full MTA service geography, which extends into parts of surrounding counties. A limitation that this dataset has compared to the primary data set is that it's based on a weekly basis instead of daily. So, the plan for primary and this secondary data set is to align them on a weekly basis. Since this data set has weekly reports, we will do this by aggregating daily MTA ridership into weekly totals.

In this section, we mapped subway complexes to taxi zones by creating a buffer radius of 0.5 miles for each Station, then tracked the ridership per zone so we could Identify the most popular stations on the weekdays and weekends.
<!-- Checked for Spelling and Grammar (Dustin, 3-18-2026) -->

````python
# Make a copy and ensure CRS
stations_gdf = subway_stations_gdf.copy()
if stations_gdf.crs is None:
    stations_gdf = stations_gdf.set_crs('EPSG:4326')

# Convert to projected CRS for buffering
stations_gdf = stations_gdf.to_crs('EPSG:3857')
taxi_zones_proj = taxi_zones_gdf.to_crs('EPSG:3857')

# Create buffers (0.1 miles - same as heatmap)
buffer_miles = 0.1
buffer_meters = buffer_miles * 1609.34
stations_gdf['buffer_geom'] = stations_gdf.geometry.buffer(buffer_meters)

# Spatial join using buffers
stations_buffer = stations_gdf[['Complex ID', 'Stop Name', 'Borough']].copy()
stations_buffer = stations_buffer.set_geometry(stations_gdf['buffer_geom'])
stations_buffer.crs = stations_gdf.crs

station_zone_mapping = gpd.sjoin(
    stations_buffer,
    taxi_zones_proj,
    how='left',
    predicate='intersects'
)

# Remove duplicates
station_zone_mapping = station_zone_mapping.drop_duplicates(subset=['Complex ID', 'locationid'])

# Prepare MTA data
mta_df['date'] = pd.to_datetime(mta_df['date'])
mta_df['station_complex_id'] = mta_df['station_complex_id'].astype(str)
station_zone_mapping['Complex ID'] = station_zone_mapping['Complex ID'].astype(str)
station_zone_mapping['locationid'] = station_zone_mapping['locationid'].astype(str)
taxi_zones_gdf['locationid'] = taxi_zones_gdf['locationid'].astype(str)

# ============================================
# MTA ANALYSIS BY TAXI ZONE
# ============================================

# Merge ridership with zone mapping for all data
mta_with_zones = station_zone_mapping.merge(
    mta_df,
    left_on='Complex ID',
    right_on='station_complex_id',
    how='left'
)

# Calculate average ridership by taxi zone for weekdays/weekends
weekday_mta_zones = mta_with_zones[mta_with_zones['weekend'] == 0].groupby('locationid').agg({
    'ridership': 'mean'
}).reset_index()

weekend_mta_zones = mta_with_zones[mta_with_zones['weekend'] == 1].groupby('locationid').agg({
    'ridership': 'mean'
}).reset_index()

weekday_mta_zones.columns = ['locationid', 'avg_ridership']
weekend_mta_zones.columns = ['locationid', 'avg_ridership']

# Merge with taxi zones
zones_weekday = taxi_zones_gdf.merge(weekday_mta_zones, on='locationid', how='left')
zones_weekday['avg_ridership'] = zones_weekday['avg_ridership'].fillna(0)
manhattan_weekday = zones_weekday[zones_weekday['borough'] == 'Manhattan']

#print(manhattan_weekday.columns)

print(f"\nWeekday - MTA Manhattan Summary (by taxi zone):")
print(f"Total avg ridership: {manhattan_weekday['avg_ridership'].sum():,.0f}")
print("\nTop 5 zones:")
top = manhattan_weekday.nlargest(5, 'avg_ridership')[['zone', 'avg_ridership','locationid']]
for idx, row in top.iterrows():
    print(f"  {row['zone']}, {row['locationid']}: {row['avg_ridership']:,.0f} avg riders")

zones_weekend = taxi_zones_gdf.merge(weekend_mta_zones, on='locationid', how='left')
zones_weekend['avg_ridership'] = zones_weekend['avg_ridership'].fillna(0)
manhattan_weekend = zones_weekend[zones_weekend['borough'] == 'Manhattan']

print(f"\nWeekend - MTA Manhattan Summary (by taxi zone):")
print(f"Total avg ridership: {manhattan_weekend['avg_ridership'].sum():,.0f}")
print("\nTop 5 zones:")
top = manhattan_weekend.nlargest(5, 'avg_ridership')[['zone', 'avg_ridership', 'locationid']]
for idx, row in top.iterrows():
    print(f"  {row['zone']}, {row['locationid']}: {row['avg_ridership']:,.0f} avg riders")

# Calculate percent change
weekday_total = manhattan_weekday['avg_ridership'].sum()
weekend_total = manhattan_weekend['avg_ridership'].sum()
pct_change = ((weekend_total - weekday_total) / weekday_total) * 100
print(f"\n MTA Weekend vs Weekday change (by zone): {pct_change:+.1f}%")

#plt.show(maps.plot_subway_heatmap_by_day(mta_df, subway_stations_gdf, taxi_zones_gdf, 6, 75000, summary=False))
#maps.plot_subway_heatmap_by_day(mta_df, subway_stations_gdf, taxi_zones_gdf, day_num=6, summary=True)
````

In order to look at the relationship between gas prices and subway ridership, we aggregated daily ridership into weekly totals and plotted them against the weekly average of regular gas prices of 2023 in New York City. Gas prices ranged from around \\$3.10 to \\$3.75 per gallon in the cleaned data. Under inspection, we found an outlier in January that corresponded with the New Year’s holiday which had really low ridership and didn't represent weekly trends, so it was removed. 

In regards to the data visualization, the scatter plot showed a very weak negative correlation with the Pearson correlation being -0.216 and the p-value of 0.127. The negative direction of the relationship rejects our notion that higher gas prices would lead to more commuters to subway stations rather than ridehailing. Another thing to note is that the p value is above the .05 threshold so the relationship between gas and ridership could be due to chance. This suggests that gas prices alone can’t explain the trends in subway ridership, and that there are other factors at play. 

Note, in this section, we separate weekdays and weekends due to the large shift in riders between weekdays and weekends. 

Rain & Snow Analysis:
From our rain analysis we see that there is a slightly significant negative correlation (r = -0.32, r = -0.14) between rain and ridership when separating the weekdays and weekends as well as removing our list of outliers. For snow we note that no snow days occurred on weekends and the correlation on the weekdays is far less impactful than rain (r = -0.09).


We take the average daily temperature and daily riders to avoid the influence of the time of day, and all the activities and habits of people surrounding that.

We hypothesized a lower threshold of 50°F (10°C) and an upper threshold of 80°F (27°C), a lower or higher temperature than those thresholds resulting in a shift of riders from subways to ride hailing. 

Overall, our original idea of ridership decreasing past 50°F and 80°F was correct, however, the data does not reflect the hypothesized thresholds. Instead, the data shows peaks; ridership and number of rides decreasing as temperature drifts from this peak. 

Peaks in ridership were around 57-63.5°F (roughly 14-17°C) for both ride-hailing and subway modes indicating that travel demand is highest in mild, comfortable temperatures where outdoor activity is abundant.

Visually, the graph for ride hailing is much steeper than the subway's. This suggests that ride hailing users are more sensitive to temperature shifts, changes in temperature leading to greater changes in rides as compared to subway riders, while subways have more regular commuters. These commuters are most likely workers as the subway experiences more riders during weekdays and less over weekends; ride hailing holds a similar amount of riders over both.

Overall, our lower threshold was the closest to the real peak and there are little visible relationships between average daily ride share trips, subway ridership, and daily temperature.

````python
#import data
daily_weather = pd.read_csv("data/01-interim/Weather/CLEANED_CentralPark_Weather2023.csv")
daily_ridership = pd.read_csv("data/01-interim/MTA_subway/MTA_Subway_Daily_Manhattan_Totals.csv")
daily_rides = pd.read_csv("data/02-processed/daily_total_ridership.csv")

daily_weather = daily_weather.drop(['Station','Name','Latitude','Longitude','Snow','Snow?', 'Precipitation', 'Rain?'], axis = 1)
daily_weather['Date'] = pd.to_datetime(daily_weather['Date'])
daily_ridership['date'] = pd.to_datetime(daily_ridership['date'])
daily_rides['date'] = pd.to_datetime(daily_rides['date'])
````

````python
#subway ridership and average temperature threshold (visual)
daily_ridership['is_weekend'] = daily_ridership['date'].dt.dayofweek > 4
merged = daily_ridership.merge(daily_weather, left_on='date', right_on='Date')
sns.lmplot(x='Avg_Temp', y='ridership', data=merged, hue='is_weekend', lowess = True, scatter_kws={'alpha':0.3}, legend = False)
plt.legend(title = 'Day Type', labels = ['Weekday',' ','Weekend',' '], loc = 'center left')
plt.title('Ridership Trend by Temperature (Subways)')
plt.xlabel('Average Temperature (°F)')
plt.ylabel('Daily Ridership (1e6)')
plt.subplots_adjust(right=0.8)
plt.tight_layout()
````

````python
#subway ridership and average temperature threshold (numerical)
#define weekday/weekends
subway_weekdays = merged[~merged['is_weekend']].copy()
subway_weekends = merged[ merged['is_weekend']].copy()

#fit lowess
lowess_weekday = lowess(subway_weekdays['ridership'], subway_weekdays['Avg_Temp'])
lowess_weekend = lowess(subway_weekends['ridership'], subway_weekends['Avg_Temp'])

#find peaks
x_wd, y_wd = lowess_weekday.T
x_we, y_we = lowess_weekend.T

peak_temp_weekday = x_wd[np.argmax(y_wd)]
peak_ridership_wd = y_wd.max()

peak_temp_weekend = x_we[np.argmax(y_we)]
peak_ridership_we = y_we.max()

print(f"Weekday subway ridership peak at ~ {peak_temp_weekday:4.1f} °F, averaging {peak_ridership_wd/1e6:5.2f} million")
print(f"Weekend subway ridership peak at ~ {peak_temp_weekend:4.1f} °F, averaging {peak_ridership_we/1e6:5.2f} million")
````

````python
#ride hailing and average temperature threshold (visual)
daily_rides['date'] = pd.to_datetime(daily_rides['date'])

daily_rides['is_weekend'] = daily_rides['date'].dt.dayofweek > 4
merged2 = daily_rides.merge(daily_weather, left_on='date', right_on='Date')
sns.lmplot(x='Avg_Temp', y='total_daily_trips', data=merged2, hue='is_weekend', lowess = True, scatter_kws={'alpha':0.3}, legend = False)
plt.legend(title = 'Day Type', labels = ['Weekday',' ','Weekend',' '], loc = 'lower left')
plt.title('Ride Trend by Temperature (Ride Hailing)')
plt.xlabel('Average Temperature (°F)')
plt.ylabel('Daily Rides (1e6)')
plt.subplots_adjust(right=0.8)
plt.tight_layout()
````

````python
#ride hailing and average temperature threshold (numerical)
#define weekday/weekends
ridership_weekdays = merged2[~merged2['is_weekend']].copy()
ridership_weekends  = merged2[ merged2['is_weekend']].copy()

#filter ends
temp_min, temp_max = 45, 90

ridership_weekdays_filtered = ridership_weekdays[(ridership_weekdays['Avg_Temp'] >= temp_min) &(ridership_weekdays['Avg_Temp'] <= temp_max)].copy()
ridership_weekends_filtered = ridership_weekends[(ridership_weekends['Avg_Temp'] >= temp_min) &(ridership_weekends['Avg_Temp'] <= temp_max)].copy()

#fit lowess
lowess_weekday2 = lowess(ridership_weekdays_filtered['total_daily_trips'], ridership_weekdays_filtered['Avg_Temp'])
lowess_weekend2 = lowess(ridership_weekends_filtered['total_daily_trips'], ridership_weekends_filtered['Avg_Temp'])

#find peaks
x_wd2, y_wd2 = lowess_weekday2.T
x_we2, y_we2 = lowess_weekend2.T

peak_temp_weekday2 = x_wd2[np.argmax(y_wd2)]
peak_ridership_wd2 = y_wd2.max()

peak_temp_weekend2 = x_we2[np.argmax(y_we2)]
peak_ridership_we2= y_we2.max()

print(f"Weekday ridehailing rides peak at ~ {peak_temp_weekday2:4.1f} °F, averaging {peak_ridership_wd2/1e6:5.2f} million")
print(f"Weekend ridehailing rides peak at ~ {peak_temp_weekend2:4.1f} °F, averaging {peak_ridership_we2/1e6:5.2f} million")
````

Outside factors, such as legacy regulations, which limit the amount of taxi medallions (licenses), play a role in this stark contrast. This is especially so, given HVFHV’s lack of limitations; their rides are considered a pre-arranged pick up rather than street hail. Knowing this, we were able to narrow our scope from the broader question of transportation competition in New York City to a more focused comparison between ride hailing demand and subway ridership. A key discovery was the TLC data’s separation of traditional taxis and app based for hire services, which made HVFHV a better proxy for Uber and Lyft than taxis alone. Lastly, we found that transportation behavior in New York City is shaped not only by consumer choice but by regulation, geography, and service type. Street hailing taxis and pre-arranged ride hailing services operate under different rules, which likely affects where and when riders choose each mode. This ultimately led to one of our main challenges: defining exactly which form of ‘ridehailing’ we wanted to study. Yellow taxis, green taxis, HVFHV, and other FHVs serve different roles in the city, and combining them into one category would blur important differences. As a result, we focused on HVFHV activity as the clearest representation of modern app-based ridehailing. Thus giving us a more precise way to examine whether the rise of services like Uber and Lyft corresponds with changes in subway ridership. 

Our TLC Ridehailing Data was extracted from the TLC database for 2023, the latest year that was available for server-end querying to reduce the dataset that would need to be downloaded (which would’ve been in the GBs). What was left after server-end querying was still too large, thus, we downloaded the raw parquet file version locally and performed local data cleaning and tidying.
Moving on to the MTA dataset, our 2023 Manhattan ridership numbers were significantly lower on the weekends. Thus, we split the dataset into weekdays and weekends and analyzed them separately. Additionally, we chose to remove outliers that were more than two standard deviations from the norm. Almost all of these outliers were public holidays.

We see strong, positive correlation between MTA and ridehailing during the weekends, contrary to the negative correlation we had initially hypothesized. During the weekend, the correlation becomes statistically weak; no correlation on Saturdays and a weak correlation on Sundays. This is likely due to differing activities such as work or nightlife affecting correlations during weekdays vs weekends. Exploring this further, by doing zonal analysis, we find that areas with most ridership and rides shift from work areas to nightlife areas depending on the day of the week. Overall, for MTA, the most utilized areas remain consistently the working areas. When looking at correlations in each zone, we find some regions are positively correlated while others are negatively correlated. During the weekdays, all zones are positively correlated while on the weekends, it is a mixed bag of correlations. We can therefore conclude our initial hypothesis for a negative correlation between subway ridership and ridehailing rides was incorrect, it is instead highly dependent on the day of the week and zone.

Another factor we were interested in was a possible relationship between gas prices and subway ridership. As gas prices rose, we anticipated less people to opt for personal vehicles and for ridehailing fares to increase accordingly. In this analysis, we aggregated daily ridership totals into weekly totals, as a result of our gas dataset being reported as weekly averages. We plotted this aggregate ridership data against the weekly average of regular gas prices of 2023 in New York City. Gas prices ranged from \\$3.10 to \\$3.75 per gallon in the cleaned data. The scatter plot showed a very weak negative correlation with the Pearson correlation (-0.216) and p-value (0.127). Due to the p-value being so high, we can reject the notion that gas prices have any correlation to subway ridership. 

When testing our hypothesis for rain and snow, we found that snow does not have a significant correlation with the subway’s ridership. However, during the weekdays, rain does have a significant (p-value: 2.57*10^-7) but weak correlation (r = -0.32) with decreasing ridership. This shows our hypothesis, that rain and snow will decrease ridership, was true for rain when limited to weekdays, and false for snow. Overall, our original idea of subway ridership decreasing when average daily temperatures decreased past 50°F and increasing above 80°F was correct, however, the data does not reflect the hypothesized thresholds and did not show a correlation between temperature and a user’s choice between subway and ridehailing. We separated weekdays and weekends once again, scatter plotted the average daily temperature and daily totals for ridership, and fitted a line in each scatterplot. As a result, the data showed peaks at around 57-63.5°F; ridership and the number of rides decreased as temperature drifted from this peak. Visually, the graph for ride hailing was much steeper than that of the subway’s. This suggests that ridehailing users are more sensitive to temperature shifts, changes in temperature leading to greater changes in the amount of rides as compared to subway riders, who were more regular commuters. These commuters are most likely workers as the subway experiences more riders during the weekdays and less over weekends, while ride hailing held a similar amount of riders in both. Overall, our lower threshold was the closest to the real peak and there are little relationships between average daily ridehailing trips, subway ridership, and daily temperatures.

Thus, overall, we found that New York has its place for both the subway and ridehailing services. While one supports regular working commuters, the other uplifts the nightlife scene; while one dominates the weekdays, the other impacts the weekends. Ultimately, people are much more complex than we initially hypothesized and cannot be predicted to make a choice on taking the subway vs ordering a ridehailing service on weather, gas prices, or preference alone. Summarizing our findings in one sentence: Subway ridership and ridehailing rides have a positive correlation, ridership and rides peak at 57-63.5°F, there is a negative effect on subway ridership when it is raining, and gas has no effect.
